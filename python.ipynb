```python
import matplotlib.pyplot as plt
import pandas as pd
from sqlalchemy import create_engine
```

```python
# Define the database connection details
db_url = 'postgresql+psycopg2://postgres:postgres@localhost:5432/postgres'
table_name = 'Movie'

# Create an engine instance
engine = create_engine(db_url)

# Fetch data from the table and load it into a DataFrame
query = f'SELECT filmtv_id, title, year, genre, duration, country, avg_vote, critics_vote, public_vote, total_votes FROM {table_name}'
movies_df = pd.read_sql(query, engine)

# Display the first few rows of the dataframe
print(movies_df.head())
```

```python
# Visualize the distribution of movie durations
plt.figure(figsize=(10, 6))
plt.hist(movies_df['duration'], bins=30, color='skyblue', edgecolor='black')
plt.title('Distribution of Movie Durations')
plt.xlabel('Duration (minutes)')
plt.ylabel('Number of Movies')
plt.grid(True)
plt.show()
```

The histogram shows that most movies have durations ranging between 80 to 120 minutes, with a noticeable peak around 90 minutes. This suggests that the majority of films in the dataset adhere to a standard feature-length duration.

```python
# Visualize the distribution of genres
genre_counts = movies_df['genre'].value_counts()

plt.figure(figsize=(12, 8))
genre_counts.plot(kind='bar', color='lightcoral', edgecolor='black')
plt.title('Distribution of Movie Genres')
plt.xlabel('Genre')
plt.ylabel('Number of Movies')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y')
plt.show()
```

The bar chart illustrates that the most common genres in the dataset are Drama, Comedy, and Documentary, indicating a preference or higher production rate for these genres compared to others.

```python
# Visualize the distribution of average votes
plt.figure(figsize=(10, 6))
plt.hist(movies_df['avg_vote'], bins=20, color='lightgreen', edgecolor='black')
plt.title('Distribution of Average Votes')
plt.xlabel('Average Vote')
plt.ylabel('Number of Movies')
plt.grid(True)
plt.show()
```

The histogram of average votes shows that most movies receive ratings between 5 and 8, with a peak around 6-7. This suggests a tendency towards moderate to slightly above-average ratings among the movies in the dataset.

```python
# Visualize the comparison between critics' votes and public votes
plt.figure(figsize=(10, 6))
plt.scatter(movies_df['critics_vote'], movies_df['public_vote'], alpha=0.6, color='purple')
plt.title('Critics\' Votes vs Public Votes')
plt.xlabel('Critics\' Vote')
plt.ylabel('Public Vote')
plt.grid(True)
plt.show()
```

The scatter plot reveals a generally positive correlation between critics' votes and public votes. However, there are instances where the two diverge significantly, indicating differing opinions between critics and the general public on certain movies.

```python
# Visualize the number of movies produced each year
movies_per_year = movies_df['year'].value_counts().sort_index()

plt.figure(figsize=(14, 8))
movies_per_year.plot(kind='line', marker='o', color='orange')
plt.title('Number of Movies Produced Each Year')
plt.xlabel('Year')
plt.ylabel('Number of Movies')
plt.grid(True)
plt.show()
```

The line chart shows fluctuations in movie production over the years, with notable peaks and troughs. These variations could correspond to historical events, technological advancements, or changes in industry trends.

```python
# Visualize the distribution of movies by country
country_counts = movies_df['country'].value_counts().head(10)  # Taking top 10 countries for better visualization

plt.figure(figsize=(12, 8))
country_counts.plot(kind='bar', color='lightblue', edgecolor='black')
plt.title('Top 10 Countries by Number of Movies Produced')
plt.xlabel('Country')
plt.ylabel('Number of Movies')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y')
plt.show()
```

The bar chart for the top 10 countries by movie production shows that the United States leads by a significant margin, followed by Italy and the United Kingdom. This highlights the dominance of certain countries in the film industry.
